```python
import numpy as np
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var_pd=pd.read_csv('./output/inference/exp5/var_at_epoch100.csv').apply(lambda x:np.sqrt(x))
mean_pd=pd.read_csv('./output/inference/exp5/mean_at_epoch100.csv')
```

```python
var_ab=np.array(var_pd[var_pd.labels==1]['var'])
mean_ab=np.array(mean_pd[mean_pd.labels==1]['scores'])
var_n=np.array(var_pd[var_pd.labels==0]['var'])
mean_n=np.array(mean_pd[mean_pd.labels==0]['scores'])
```

```python
abn_mean = mean_pd.loc[mean_pd.labels == 1]['scores']
nrm_mean = mean_pd.loc[mean_pd.labels == 0]['scores']
```

```python
sns.distplot(nrm_mean, label=r'Abnormal Score')
sns.distplot(abn_mean, label=r'Normal Score')
```

```python
sns.distplot(nrm_mean, label=r'Abnormal Score')
sns.distplot(abn_mean, label=r'Normal Score')
# sns.scatterplot(mean_n,var_n*200,label='normal std', marker='.')
# sns.scatterplot(mean_ab,var_ab*200,label='abnormal std', marker='.')
```

```python
plt.scatter(mean_ab,var_ab,label='abnormal', marker='.')
plt.scatter(mean_n,var_n,label='normal', marker='.')
# plt.xlim(0, 0.8)
# plt.ylim(0, 0.15)
plt.title('mean and std')
plt.xlabel('mean')
plt.ylabel('std')
plt.legend(loc='best')
```

```python
plt.scatter(mean_n,var_n,label='normal', marker='.')
plt.scatter(mean_ab,var_ab,label='abnormal', marker='.')

# plt.xlim(0, 0.8)
# plt.ylim(0, 0.15)
plt.title('mean and std')
plt.xlabel('mean')
plt.ylabel('std')
plt.legend(loc='best')
```
